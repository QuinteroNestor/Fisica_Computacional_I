# Ajuste de datos con Numpy: Uso de `polyfit()`

### Ejemplo: Ajuste (fit) lineal ($y=mx+b$) de datos

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos experimentales (puntos x, y)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.5, 3.5, 4.5, 5.5, 6.5])

# Realizar el ajuste lineal usando la función polyfit de NumPy
coefficients = np.polyfit(x, y, 1)  # Ajuste lineal de grado 1 (una línea recta)
#Es un array que contiene los coeficientes del polinomio en orden descendente de potencias de x.

m = coefficients[0]
b = coefficients[1]

# Generar la curva ajustada (fit) 
y_fit = m * x + b

# Graficar los datos experimentales y la línea de ajuste
plt.scatter(x, y, label='Datos experimentales')
plt.plot(x, y_fit, color='red', label='Ajuste lineal')

# Agregar etiquetas y título
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ajuste lineal de datos experimentales')

# Agregar una leyenda
plt.legend()

# Mostrar el gráfico
plt.show()

# Imprimir los coeficientes de la línea de ajuste
print("Coeficientes del ajuste:")
print(f"m = {coefficients[0]}")
print(f"b = {coefficients[1]}")
```

### Ejemplo: Ajuste cuadrático ($y=ax^2 + bx +c$)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios con comportamiento cuadrático ax^2 + bx + c
np.random.seed(0)
x = np.random.uniform(-10, 10, 100)
y = 2 * x**2 - 3 * x + 1
y_random = y + np.random.normal(0, 5, size=len(x))

# Ajustar los datos a una función polinomial de grado 2 (función cuadrática)
coefficients = np.polyfit(x, y_random, 2)

# Generar la curva ajustada
x_fit = np.linspace(-10, 10, 100)
y_fit = np.polyval(coefficients, x_fit)
"""La función np.polyval() toma los coeficientes del polinomio y los valores de x, y 
devuelve los valores correspondientes de 'y' evaluados en el polinomio."""

# Visualizar los datos y el ajuste
plt.scatter(x, y_random, label='Datos experimentales')
plt.plot(x_fit, y_fit, color='red', label='Ajuste cuadrático')

# Personalizar la figura
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ajuste de datos a una función cuadrática')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir los coeficientes del ajuste
print("Coeficientes del ajuste:")
print("a =", coefficients[0])
print("b =", coefficients[1])
print("c =", coefficients[2])
```
