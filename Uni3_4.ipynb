## <font color='blue'> Gráficas de contorno en Matplotlib  </font>

---

### Definición

Los **gráficos de contorno**, también conocidos como **mapas de contorno**, son representaciones visuales de 
*funciones bidimensionales*. Muestran las curvas de nivel que conectan puntos de igual valor de la función. 
Son útiles para visualizar patrones y tendencias en datos bidimensionales, como campos de temperatura, presión, 
densidad, entre otros.

Hay dos funciones de **Matplotlib** que pueden resultar útiles para esta tarea: `plt.contour()` para trazados de contornos y 
`plt.contourf()` para trazados de contornos rellenos. En este sección se analizan varios ejemplos de su uso. 

### Características
- Visualiza funciones bidimensionales.
- Representa curvas de nivel.
- Puede personalizarse con diferentes colores, estilos de línea y niveles de contorno.

### Uso

Los gráficos de contorno son ampliamente utilizados en campos como la física, la meteorología, la geografía, 
la ingeniería y otras disciplinas científicas para visualizar y analizar datos bidimensionales.

### Sintaxis básica de `plt.contour()` y `plt.contourf()`
    
`plt.contour()` y `plt.contourf()` son dos funciones de Matplotlib que se utilizan para crear gráficos de contorno. 
La sintaxis básica de ambas funciones es similar, pero hay algunas diferencias clave.

#### `plt.contour()`

- `plt.contour()` crea líneas de contorno que conectan puntos de igual valor en los datos.
- Las líneas de contorno no están rellenas, es decir, el área entre las líneas no se colorea.
- Es útil para visualizar los contornos de los datos, pero no proporciona una representación visual completa de la distribución de los datos.
- La función devuelve un objeto de contorno que puede ser utilizado para agregar una barra de color mediante `plt.colorbar()`.
    
```python
import matplotlib.pyplot as plt

plt.contour(X, Y, Z, levels=None, cmap='viridis', linewidth=2, corner_mask=True)

```

**Parámetros:**

- X: Matriz 2D o vector 1D que representa las coordenadas X de los datos.
- Y: Matriz 2D o vector 1D que representa las coordenadas Y de los datos.
- Z: Matriz 2D que representa los valores de los datos a plotear.

(**IMPORTANTE**: X, Y y Z deben tener las mismas dimensiones y ser matrices 2D de NumPy que representan la
cuadrícula de datos y los valores de los datos, respectivamente.)

- levels: (Opcional) Vector que especifica los niveles de contorno a plotear. Si no se especifica, se utilizan niveles predeterminados.
- cmap: (Opcional) Nombre de la paleta de colores a utilizar. La paleta de colores predeterminada es 'viridis'.
- linewidth: (Opcional) Grosor de las líneas de contorno. El valor predeterminado es 2.
- corner_mask: (Opcional) Si es True, se elimina el contorno alrededor de los bordes de la gráfica. 
El valor predeterminado es True.

#### `plt.contourf()`

- `plt.contourf()` también crea líneas de contorno, pero además rellena el área entre las líneas con colores.
- El área entre las líneas de contorno se colorea de acuerdo a una escala de colores, lo que proporciona una 
representación visual completa de la distribución de los datos.
- Es útil cuando se desea resaltar la distribución de los datos y visualizar áreas de diferentes niveles de valores.
- La función también devuelve un objeto de contorno que puede ser utilizado para agregar una barra de color 
    mediante `plt.colorbar()`.

```python
import matplotlib.pyplot as plt

plt.contourf(X, Y, Z, levels=None, cmap='viridis')


```

**Parámetros:**

- X: Matriz 2D o vector 1D que representa las coordenadas X de los datos.
- Y: Matriz 2D o vector 1D que representa las coordenadas Y de los datos.
- Z: Matriz 2D que representa los valores de los datos a plotear.
- levels: (Opcional) Vector que especifica los niveles de contorno a plotear. Si no se especifica, 
se utilizan niveles predeterminados.
- cmap: (Opcional) Nombre de la paleta de colores a utilizar. La paleta de colores predeterminada es 'viridis'.


#### Diferencias clave:

- `plt.contour()` crea líneas de contorno, mientras que `plt.contourf()` rellena las áreas entre las líneas de contorno con colores.
- `plt.contourf()` no tiene un parámetro linewidth como `plt.contour()`.
- La elección entre estas dos funciones depende de la naturaleza de los datos y del tipo de visualización deseada.

### Función `np.meshgrid(x, y)`

La función `np.meshgrid(x, y)` en **NumPy** se utiliza para crear una cuadrícula de coordenadas bidimensionales a 
partir de dos vectores unidimensionales de coordenadas x e y. Esta función es útil para generar los puntos 
de la cuadrícula necesarios para evaluar funciones bidimensionales en un espacio 2D, como `plt.contour()` y `plt.contourf()`.

Cuando se llama con dos vectores x e y, `np.meshgrid(x, y)` devuelve dos matrices X e Y que contienen las coordenadas x e y 
de cada punto en la cuadrícula, respectivamente.

En general, se recomienda usar `np.meshgrid(x,y)` con `plt.contour()` y `plt.contourf()` a menos que haya una razón 
específica para no hacerlo. Esto hace que el código sea más claro y legible, y también puede ayudar a evitar errores.

---

### Función `plt.colorbar()`

-`plt.colorbar()` es una función en **Matplotlib** que se utiliza para agregar una barra de color a un gráfico. 
Esta barra de color proporciona una referencia visual para la relación entre los valores de los datos y 
los colores utilizados en el gráfico. 

-Es especialmente útil cuando se trabaja con gráficos de contorno, gráficos de dispersión 3D,
mapas de calor u otros tipos de gráficos que utilizan colores para representar valores de datos.

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Definimos algunas funciones f(x,y) para ser usadas

def f_1(x, y):
    return np.sin(X) + np.cos(Y)
    
def f_2(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f_1(X,Y)
W = f_2(X,Y)

# Crear una figura y dos subgráficas en una fila
plt.figure(figsize=(8, 5))  # Tamaño de la figura (dpi=100)
#-----------------------------------
plt.subplot(1, 2, 1)  # Subgráfica 1
plt.contour(X, Y, Z, 10, cmap='RdGy') 
plt.colorbar() 
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contorno: plt.contour')

#-----------------------------------
plt.subplot(1, 2, 2)  # Subgráfica 2
plt.contourf(X, Y, Z, cmap='coolwarm')
plt.colorbar() 
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contorno: plt.contourf')

#----------------------------------
plt.tight_layout()  # Ajustar el espaciado
plt.show()
````

````python
# Define the temperature function
def T(x, y):
    return np.exp(-(x**2 + y**2) / 2)  # Example Gaussian heat distribution

# Generate coordinates for the mesh
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Create the contour plot
plt.contourf(X, Y, T(X, Y), cmap='hot')

# Configure the plot
plt.colorbar(label='Temperature ($^{\circ}$C)')
plt.title('Two-Dimensional Heat Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-2, 2)
plt.ylim(-2, 2)


plt.show()
````

````python
# Constantes físicas
q = 1e-9  # Carga eléctrica (en Coulombs)
a = 0.5  # Distancia de separación entre las cargas (en metros)
epsilon_0 = 8.854e-12  # Permitividad del vacío (en F/m)

# Definir la función para calcular el potencial eléctrico
def phi(x, y):
    r2 = (x**2 + y**2)
    return q * a / (4 * np.pi * epsilon_0 * r2)

# Generar coordenadas de la malla
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

# Crear el gráfico de contorno relleno
plt.contourf(X, Y, phi(X, Y), cmap='summer')

# Configurar el gráfico
plt.colorbar(label='Potencial eléctrico (V)')

plt.title('Potencial de un dipolo eléctrico')
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.xlim(-0.1,0.1)
plt.ylim(-0.1,0.1)

plt.show()
````

#### Ejemplo: Superficies Equipotenciales y líneas de campo E de un par de cargas

````python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular el potencial eléctrico y el campo eléctrico de un par de cargas en una cuadrícula
def electric_field_potential(x, y, q, d):
    # Constante de permitividad del vacío
    epsilon_0 = 8.854e-12
    
    # Distancias desde las cargas
    r1 = np.sqrt((x - d/2)**2 + y**2)
    r2 = np.sqrt((x + d/2)**2 + y**2)
    
    # Potencial eléctrico debido a cada carga
    V1 = q / (4 * np.pi * epsilon_0 * r1)
    V2 = -q / (4 * np.pi * epsilon_0 * r2)
    
    # Suma del potencial eléctrico debido a ambas cargas
    V = V1 + V2
    
    # Componentes del campo eléctrico debido a cada carga
    Ex1 = q / (4 * np.pi * epsilon_0 * r1**2) * (x - d/2) / r1**2
    Ey1 = q / (4 * np.pi * epsilon_0 * r1**2) * y / r1**2
    
    Ex2 = q / (4 * np.pi * epsilon_0 * r2**2) * (x + d/2) / r2**2
    Ey2 = q / (4 * np.pi * epsilon_0 * r2**2) * y / r2**2
    
    # Suma de los campos eléctricos de ambas cargas
    Ex = Ex1 + Ex2
    Ey = Ey1 + Ey2
    
    return V, Ex, Ey

# Parámetros del sistema (cargas y distancia entre ellas)
q = 1.0e-9   # Magnitud de la carga
d = 1.0   # Distancia entre las cargas

# Crear una cuadrícula meshgrid
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)

# Calcular el potencial eléctrico y el campo eléctrico del sistema en la cuadrícula
V, Ex, Ey = electric_field_potential(X, Y, q, d)

# Crear la figura y los ejes
plt.figure(figsize=(10, 5))

# Trazar las líneas de campo eléctrico
plt.contour(X, Y, V, 300, cmap='viridis')  # Superficies equipotenciales
plt.streamplot(X, Y, Ex, Ey, color='k', linewidth=1, density=2)  # Líneas de campo eléctrico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Superficies Equipotenciales y Líneas de Campo Eléctrico')
plt.colorbar(label='Potencial eléctrico (V)')



plt.tight_layout()
plt.show()
````
