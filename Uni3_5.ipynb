## <font color='blue'> Gráficas 3D: Visualización de Datos en 3D con Matplotlib  </font>


*Matplotlib* proporciona herramientas para crear **gráficos en tres dimensiones** (**3D**) utilizando el módulo `mpl_toolkits.mplot3d`. 
Estos gráficos son útiles para visualizar datos que tienen tres variables o dimensiones.

- Los **gráficos 3D** en *Matplotlib* permiten la visualización de datos con tres dimensiones, proporcionando una perspectiva 
más rica y detallada en comparación con los gráficos 2D. 
- Estos gráficos son especialmente útiles en campos como la física, la ingeniería y las ciencias naturales, donde los 
datos a menudo tienen múltiples dimensiones.

---


## <font color='blue'>  Importar el Módulo Necesario </font> 

---

Para crear gráficos en 3D, primero debes importar el **módulo Axes3D** de `mpl_toolkits.mplot3d` junto con Matplotlib:

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

## <font color='blue'>  Crear una Figura 3D </font> 

---


Para crear una figura en 3D, utilizas `plt.figure()` y añades un subplot 3D con el método `add_subplot()` 
especificando el argumento `projection='3d'`. Este argumento adicional se usa para especificar que el subplot debe ser 
tridimensional. 

(111) significa que estamos creando una cuadrícula de 1x1 y añadiendo el primer subplot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

#### Alternativamente

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

## <font color='blue'>  Personalización de Gráficos 3D </font> 

---

Puedes personalizar los gráficos 3D de varias maneras:

- **Etiquetas de los ejes**: ax.set_xlabel('X'), ax.set_ylabel('Y'), ax.set_zlabel('Z')
- **Título**: ax.set_title('Título del Gráfico')
- **Mapas de Colores**: Utiliza el parámetro cmap para cambiar el mapa de colores.
- **Barras de Color**: Añade una barra de color con fig.colorbar(surf).

## <font color='blue'> Tipos de gráficos </font> 

### Gráficos de líneas en 3D: `plot3D()`
---

- El gráfico tridimensional más básico es una línea creado a partir de conjuntos de tripletas $(x, y, z)$. 
- El método `plot3D(x,y,z)` en Matplotlib se refiere generalmente a la función plot en el módulo `mpl_toolkits.mplot3d`, 
que permite graficar líneas en 3D.

#### Ejemplo: Espiral trigonométrica

```python
# Crear una figura
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Datos para una línea tridimensional
z = np.linspace(0, 15, 1000)
x = np.sin(z)
y = np.cos(z)
ax.plot3D(x, y, z, '-b')

# Añadir etiquetas a los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Añadir un título
ax.set_title('Línea 3D básica')

# Mostrar el gráfico
plt.show()
```

#### Ejemplo: Trayectoria de un electrón en un campo magnético

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parámetros
q = -1.6e-19  # Carga del electrón (Coulombs)
m = 9.11e-31  # Masa del electrón (kg)
B = 1e-3      # Campo magnético (Tesla)
v0 = 1e6      # Velocidad inicial (m/s)
t = np.linspace(0, 1e-6, 1000)  # Tiempo (segundos)

# Ecuaciones del movimiento
omega = q * B / m  # Frecuencia angular
x = v0 / omega * np.sin(omega * t)
y = v0 / omega * (1 - np.cos(omega * t))
z = np.zeros_like(t)  # Asumimos que no hay movimiento en z

# Crear una figura
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la trayectoria 3D
ax.plot3D(x, y, z, 'red')

# Añadir etiquetas a los ejes
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Z (m)')

# Añadir un título
ax.set_title('Trayectoria de un electrón en un campo magnético')

# Mostrar el gráfico
plt.show()
```

### Gráficos de Dispersión en 3D: `scatter3D()`
---

Puedes crear gráficos de dispersión en 3D utilizando el método `scatter3D()` del objeto Axes3D

```python
# Datos de ejemplo
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gráfico de dispersión
ax.scatter3D(x, y, z)

# Etiquetas de los ejes
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('Gráfico de Dispersión')

# Mostrar el gráfico
plt.show()
```

### Gráficos de Superficie en 3D: `plot_surface()`

---
Para crear gráficos de superficie en 3D, puedes utilizar el método : `plot_surface()` del objeto Axes3D. `plot_surface(X, Y, Z, cmap='viridis')`: Método para crear gráficos 
de superficie tridimensionales en Matplotlib. **cmap** efine el mapa de colores utilizado para colorear la superficie.

```python
# Datos de ejemplo
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Gráfico de superficie
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Añadir una barra de color
#fig.colorbar(surf)

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar el gráfico
plt.show()
```

### Gráficos de Malla 3D: `plot_wireframe()`

---

Utilizados para representar superficies como una malla en lugar de una superficie sólida.

`plot_wireframe(X,Y,Z)` es un método del objeto **Axes3DSubplot** que se utiliza para crear un gráfico de malla tridimensional. 
Este método dibuja una superficie en 3D representada por líneas en las direcciones X, Y, y Z.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(X) * np.sin(Y)

# Crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='lightblue')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Gráfico de Malla 3D')

plt.show()
```

#### Ejemplo 1: Superficie Potencial de un Oscilador Armónico 2D

En este ejemplo, visualizaremos la superficie potencial de un oscilador armónico en dos dimensiones. La energía potencial 
$V(x,y)$ de un oscilador armónico en 2D se puede describir como:

$$ V(x,y)= \dfrac{1}{2} (x^2 + y^2) $$

donde $k$ es la constante del resorte. 

```python
# Definimos la función potencial
def V(x, y):
    return (1/2)*k*(x**2 + y**2)
    
# Constante del resorte
k = 1

# Datos de ejemplo
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = V(X,Y)

# Crear una figura 3D
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(111, projection='3d')

# Gráfico de superficie
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Añadir una barra de color
#fig.colorbar(surf)

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Crear una figura
fig = plt.figure(figsize=(14, 6))

# Crear el primer subplot para el gráfico 3D
ax1 = fig.add_subplot(121, projection='3d')

# Generar datos de ejemplo
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)  # Plano z=0

# Constante de carga
Q = 1

# Calcular el potencial electrostático
V = Q / np.sqrt(X**2 + Y**2 + Z**2)

# Crear un gráfico de superficie 3D
surf = ax1.plot_surface(X, Y, V, cmap='viridis')

# Añadir una barra de color
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

# Añadir etiquetas a los ejes
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Potencial V')
ax1.zaxis.labelpad = 20  # Ajustar el padding de la etiqueta

# Añadir un título
ax1.set_title('Potencial Electrostático (3D)')

# Crear el segundo subplot para el gráfico de contorno
ax2 = fig.add_subplot(122)

# Crear un gráfico de contorno relleno
contour = ax2.contourf(X, Y, V, cmap='viridis')

# Añadir una barra de color
fig.colorbar(contour, ax=ax2, shrink=0.5, aspect=5)

# Añadir etiquetas a los ejes
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

# Añadir un título
ax2.set_title('Potencial Electrostático (Contorno)')

# Mostrar el gráfico
plt.show()
```
