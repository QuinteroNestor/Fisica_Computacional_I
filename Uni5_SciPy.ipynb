## <font color='blue'> 1. Integración numérica: </font> `scipy.integrate`

`scipy.integrate` es un submódulo de la biblioteca SciPy que proporciona una variedad de herramientas para realizar integración numérica, 
incluyendo la integración de funciones unidimensionales, la integración de múltiples dimensiones y la resolución de ecuaciones
diferenciales ordinarias. 

Módulos principales de `scipy.integrate`:

- **quad:** Integración de una función unidimensional en un intervalo finito o infinito.
- **dblquad:** Integración de una función en un dominio bidimensional.
- **tplquad:** Integración de una función en un dominio tridimensional.
- **romberg:** Integración de una función unidimensional usando el método de Romberg.
- **fixed_quad:** Cuadratura de una función unidimensional usando una cuadratura de Gauss fija.
- **quadrature:** Integración usando cuadratura adaptativa.
- **simps:** Integración de datos usando la regla de Simpson.
- **trapz:** Integración de datos usando la regla del trapecio.

### Ejemplo: Calcular la integral de la función $f(x) = x^2$ entre 0 y 1:

````python
from scipy.integrate import quad

# Definir la función a integrar
def f(x):
  return x**2
# Realizar la integración en el intervalo [0,1]
resultado = quad(f, 0, 1)

print(f"Resultado de la integración: {resultado[0]}")
print(f"Error estimado: {resultado[1]}")
print(resultado)
````

**IMPORTANTE:** La función `quad` devuelve una tupla con dos elementos:
1. El valor de la integral.
2. Una estimación del error absoluto en el valor de la integral.

### Ejemplo: Calcular la integral de la función $f(x) = e^{-x^2}$ entre 0 y $\infty$:

````python
import numpy as np
from scipy.integrate import quad

# Definir la función a integrar
def f_exp(x):
    return np.exp(-x**2)

# Realizar la integración en el intervalo [0, ∞)
result, error = quad(f_exp, 0, np.inf)

print(f"Resultado de la integración: {result}")
print(f"Error estimado: {error}")
````

### Ejemplo: Integrales dobles

````python
from scipy.integrate import dblquad

def f(x, y):
  return x**2 + y**2

a = 0
b = 1
c = 0
d = 1

resultado, _ = dblquad(f, a, b, c, d)
print(resultado)
````

````python
from scipy.integrate import dblquad

# Definir la función a integrar
def integrand(y, x):
    return np.exp(-x*y)

# Límites de integración para x e y
x_lower = 0
x_upper = 1
y_lower = 0
y_upper = 1

# Realizar la integración doble
result, error = dblquad(integrand, x_lower, x_upper, y_lower, y_upper)

print(f"Resultado de la integración doble: {result}")
print(f"Error estimado: {error}")
````

## <font color='blue'> 2. Solución numérica de Ecuaciones Diferenciales: </font> `solve_ivp`

`solve_ivp` es una función en Python de la biblioteca SciPy que se utiliza para resolver **ecuaciones diferenciales ordinarias** 
(EDOs) de primer orden con condiciones iniciales. 
Es una herramienta versátil y poderosa que admite diferentes métodos de integración, permitiendo modelar y analizar sistemas dinámicos en diversos campos como 
la física, la ingeniería y las ciencias en general.

### Características principales:

- **Soporte para EDOs de primer orden:** Puede resolver sistemas de EDOs de primer orden con una o más variables dependientes.
- **Métodos de integración explícitos e implícitos:** Ofrece una variedad de métodos de integración numérica, como el método de Euler explícito, el método de Runge-Kutta y el método de Backward Differentiation Formula (BDF).
- **Manejo de condiciones iniciales:** Permite especificar condiciones iniciales para cada variable dependiente al inicio del intervalo de integración.
- **Control de errores y tolerancia:** Permite ajustar la precisión de la solución y controlar el error de integración.
- **Salida flexible:** Devuelve una solución que incluye los valores de las variables dependientes en puntos específicos del intervalo de integración.


### Métodos de integración:

`solve_ivp` ofrece varios métodos de integración numérica para resolver EDOs:

- **Método de Euler explícito:** Es un método simple y rápido, pero puede ser menos preciso para problemas complejos.
- **Métodos de Runge-Kutta (RK):** Ofrecen mayor precisión que el método de Euler explícito, pero son más computacionalmente 
costosos.
- **Métodos de Backward Differentiation Formula (BDF):** Son eficientes para problemas con derivadas de alto orden y condiciones 
iniciales sensibles.
- **Método Radau:** Método implícito de Runge-Kutta de Radau IIA de orden 5.
- **Método LSODA:** Método que alterna entre un método de Adams y un método BDF dependiendo de la rigidez del problema.

El método predeterminado es el método RK de orden 5(4): `method='RK45'`

### Elegir el método de integración adecuado:

La elección del método de integración depende de:

- La complejidad de la EDO: Para problemas simples, el método de Euler explícito puede ser suficiente. Para problemas más 
    complejos, se recomiendan métodos de Runge-Kutta o BDF.
- La precisión requerida: Si se necesita alta precisión, se deben usar métodos más sofisticados como RK45 o BDF.
- El tiempo de computación disponible: Si el tiempo de computación es limitado, se pueden usar métodos más simples como el de Euler explícito.

### Ejemplo 1: Decaimiento Exponencial
---

El decaimiento exponencial es uno de los modelos más simples y comunes en física y otras ciencias. La ecuación diferencial que describe el 
decaimiento exponencial es:
$$ \large
\dfrac{dy}{dt} = -ky 
$$

donde $\large y(t)$ es función del tiempo, y $\large k$ es una constante de decaimiento. La solución exacta es:
$$ \boxed{\large y(t) = y_0 e^{-kt}} $$
donde $\large y_0$ es la condición inicial $\large y(0)$.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 1. Definir la ecuación diferencial
def exp_decay(t, y, k):
    dydt = -k * y
    return dydt

k = 0.5  # Constante de decaimiento
y0 = [1.0] # Condición inicial para y en t=0.
t_span = (0, 10)  # Intervalo de tiempo para la integración (inicio y fin)
t_eval = np.linspace(0, 10, 100)  # Puntos de tiempo donde se evaluará la solución

# Resolver la ecuación diferencial numéricamente
sol = solve_ivp(exp_decay, t_span, y0, args=(k,), method='RK45', t_eval=t_eval)

# Solución exacta
y_exact = y0[0] * np.exp(-k * t_eval)

# Graficar los resultados
plt.plot(sol.t, sol.y[0], 'b-', label='Solución numérica')
plt.plot(t_eval, y_exact, 'r--', label='Solución exacta')
plt.xlabel('Tiempo (s)')
plt.ylabel('y')
plt.title('Decaimiento Exponencial')
plt.legend()
plt.show()
````

### Ejemplo: Crecimimiento logístico
---

El crecimiento logístico es otro modelo simple utilizado para describir el crecimiento poblacional limitado por la capacidad de carga del ambiente. La ecuación diferencial que describe el crecimiento logístico es:

$$\large\frac{dy}{dt} = r y \left(1 - \frac{y}{K}\right) $$

donde $\large r$ es la tasa de crecimiento y $\large K$ es la capacidad de carga.

````python
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definir la ecuación diferencial
def logistic_growth(t, y, r, K):
    dydt = r * y * (1 - y / K)
    return dydt

# Parámetros y condiciones iniciales
r = 0.1
K = 10.0
y0 = [0.5]
t_span = (0, 100)
t_eval = np.linspace(0, 100, 200)

# Resolver la ecuación diferencial
sol = solve_ivp(logistic_growth, t_span, y0, args=(r, K), method='RK45', t_eval=t_eval)

# Graficar los resultados
plt.plot(sol.t, sol.y[0], label='y(t)')
plt.xlabel('Tiempo')
plt.ylabel('Población')
plt.title('Crecimiento Logístico')
plt.legend()
plt.show()
````

### Ejemplo 2: Oscilador armónico simple
---

El oscilador armónico simple está descrito por la ecuación diferencial:
$$\large \frac{d^2x}{dt^2} + \omega^2 x = 0 $$
donde $\large x(t)$ es la posición de la masa en el tiempo $\large t$ y $\large \omega$ es la frecuencia angular del oscilador. 
Esta ecuación se puede reformular como un sistema de dos 
ecuaciones diferenciales de primer orden:
$$\large
\frac{dx}{dt} = v 
$$
$$ \large
\frac{dv}{dt} = -\omega^2 x 
$$
La solución exacta del oscilador armónico simple es: $\boxed{\large x(t) = x_0 \cos(\omega t)},$
donde $\large x_0=x(0)$.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 1. Definir el sistema de EDOs (oscilador armónico simple)
def harmonic_oscillator(t, y, omega):
    x, v = y # Desempaquetar el vector de estado (variables del estado)
    dxdt = v
    dvdt = -omega**2 * x
    return [dxdt, dvdt]

# 2. Condiciones iniciales: posición inicial x0 y velocidad inicial v0
x0 = 1.0
v0 = 0.0
y0 = [x0, v0] # Vector de condiciones iniciales

# 3. Parámetros del oscilador
omega = 1.0  # frecuencia angular

# 4. Intervalo de tiempo para la integración
t_span = (0, 10)  # de 0 a 10 segundos
t_eval = np.linspace(0, 10, 100)  # puntos donde evaluar la solución

# 5. Resolver el sistema
sol = solve_ivp(harmonic_oscillator, t_span, y0, args=(omega,), method='RK45', t_eval=t_eval)

# 6. Solución exacta para comparación
exact_solution = y0[0] * np.cos(omega * t_eval)

# 7. Graficar los resultados
plt.figure(figsize=(8, 4))

# Posición en función del tiempo
plt.subplot(1, 2, 1)
plt.plot(sol.t, sol.y[0], 'b-', label='Solución numérica')
plt.plot(t_eval, exact_solution, 'r--', label='Solución Exacta')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (x)')
plt.title('Oscilador Armónico Simple')
plt.legend()

# Velocidad en función del tiempo
plt.subplot(1, 2, 2)
plt.plot(sol.t, sol.y[1], label='Velocidad (v)', color='r')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (v)')
plt.title('Oscilador Armónico Simple - Velocidad')
plt.legend()

plt.tight_layout()
plt.show()
````

### Ejemplo: Péndulo simple (sin fricción)
---

La ecuación del péndulo simple sin fricción es:
$$\large \frac{d^2\theta}{dt^2} + \frac{g}{L} \sin(\theta) = 0$$

donde:$\large\theta$ es el ángulo que forma el péndulo con la vertical, $\large g$ es la aceleración debida a la gravedad, 
$\large L$ es la longitud del péndulo.

Para resolver esta ecuación usando `solve_ivp`, la reescribimos como un sistema de ecuaciones de primer orden:

$$\large \frac{d\theta}{dt} = \omega$$
$$\large \frac{d\omega}{dt} = -\frac{g}{L} \sin(\theta) $$

Para ángulos pequeños $(\large \theta \approx \sin(\theta))$, la ecuación del péndulo se aproxima a:
$$\large \theta(t) = \theta_0 \cos(\sqrt{g/L} \cdot t) $$


````python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Configuración de Matplotlib para usar LaTeX
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

# Definir el sistema de EDOs para el péndulo simple
def pendulum(t, y, g, L):
    theta, omega = y  # Desempaquetar el vector de estado
    dtheta_dt = omega
    domega_dt = - (g / L) * np.sin(theta)
    return [dtheta_dt, domega_dt]

# Parámetros del péndulo
g = 9.81  # Aceleración debida a la gravedad (m/s^2)
L = 1.0   # Longitud del péndulo (m)
theta0 = np.pi / 4  # Ángulo inicial (45 grados)
omega0 = 0.0        # Velocidad angular inicial
y0 = [theta0, omega0]  # Condiciones iniciales

# Intervalo de tiempo para la integración
t_span = (0, 10)
t_eval = np.linspace(0, 10, 300)  # Puntos donde se evaluará la solución

# Resolver el sistema de EDOs
sol = solve_ivp(pendulum, t_span, y0, args=(g, L), t_eval=t_eval, method='RK45')

# Solución exacta para ángulos pequeños (aproximación)
theta_exact = theta0 * np.cos(np.sqrt(g / L) * t_eval)

# Graficar los resultados
plt.figure(figsize=(8, 4))
plt.plot(sol.t, sol.y[0], label='Solución Numérica')
plt.plot(t_eval, theta_exact, '--', label='Solución Exacta (aproximación)')
plt.xlabel('Tiempo (s)', fontsize=16)
plt.ylabel(r'$\theta$ (rad)', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend()
plt.title('Péndulo Simple (sin fricción)', fontsize=14)
plt.grid(True)
plt.show()
````
