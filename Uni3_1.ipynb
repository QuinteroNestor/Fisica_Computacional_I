---
Las **bibliotecas de visualización** en *Python* son herramientas poderosas que permiten crear gráficos 
y visualizaciones de datos de manera efectiva y atractiva. 
Algunas de las bibliotecas de visualización más populares en *Python*:

1. **Matplotlib**: Es una de las bibliotecas de visualización más utilizadas en *Python*. 
Proporciona una amplia gama de funcionalidades para crear gráficos estáticos, incluyendo gráficos de líneas, 
histogramas, gráficos de dispersión, gráficos de barras, entre otros. Matplotlib es altamente personalizable 
y permite crear visualizaciones de alta calidad. Además, es la base de muchas otras bibliotecas de visualización en *Python*.

2. **Seaborn**: Construida sobre Matplotlib, Seaborn ofrece una interfaz de alto nivel para crear gráficos 
estadísticos atractivos y informativos. Seaborn simplifica la creación de visualizaciones complejas como 
diagramas de violín, gráficos de dispersión con líneas de regresión, matrices de correlación, y más. 
Además, Seaborn tiene estilos predefinidos que mejoran la apariencia de los gráficos.

3. **Pandas**: Aunque principalmente es una biblioteca para análisis de datos, Pandas también 
proporciona funcionalidades para visualizar datos de manera rápida y sencilla. 
La función `plot()` de Pandas permite crear gráficos básicos directamente desde los DataFrames y Series de Pandas, l
o que resulta útil para visualizar rápidamente los datos durante el análisis exploratorio.

4. **Plotly**: Plotly es una biblioteca de visualización interactiva que permite crear gráficos interactivos y dinámicos. 
Ofrece una variedad de tipos de gráficos, incluyendo gráficos de barras, líneas, dispersión, superficies 3D, entre otros. Además, Plotly permite crear visualizaciones interactivas como gráficos con zoom, panning y hover.

5. **Bokeh**: Similar a Plotly, Bokeh es otra biblioteca de visualización interactiva que permite crear gráficos interactivos y dinámicos en Python. Bokeh es particularmente útil para visualizaciones de grandes conjuntos de datos y para crear aplicaciones web interactivas basadas en Python.

### <font color='red'> Ejemplo: </font> Gráfica de una línea recta $y=mx+b$

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar valores de x
x = np.linspace(0, 5, 100)  # Valores de x desde 0 hasta 5

"""
np.linspace(x_min, x_max, N) genera una secuencia de N números espaciados uniformemente entre x_min y x_max (inclusivos). 
Estos números se utilizan comúnmente como valores de x en gráficos para representar un rango de valores a lo largo del eje x. 
"""

# Definimos la ecuación de la línea y = m x + b
y = 2 * x + 1

# Crear una figura 
fig = plt.figure()  # Crear una nueva figura

# Crear la gráfica de la línea recta
plt.plot(x, y)

# Personalizar la visualización
plt.xlabel('Eje X')  # Etiqueta del eje X
plt.ylabel('Eje Y')  # Etiqueta del eje Y
plt.title('Gráfica de una línea recta')  # Título del gráfico
plt.grid(True)  # Activar la cuadrícula en el gráfico

# Mostrar la gráfica
plt.show()
````

### <font color='red'> Ejemplo: </font> Varias gráficas en una sola figura

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar valores de x
x = np.linspace(0, 4, 100)  # Valores de x desde 0 hasta 5

# Crear una figura 
fig=plt.figure(figsize=(6, 3))  # Ancho: 6 in, Altura: 3 in

# Crear la gráfica de la línea recta
plt.plot(x, 2*x + 1, color='red',linestyle='-', label='$2x+1$')
plt.plot(x, 2*x + 2, color='blue',linestyle='--', label='$2x+2$')
plt.plot(x, x**2 + 2*x + 1, color='green',linestyle=':', label='$x^2+2x+1$')

# Personalizar la visualización
plt.xlabel('Eje X')  # Etiqueta del eje X
plt.ylabel('Eje Y')  # Etiqueta del eje Y
plt.title('Varias funciones en una gráfica')  # Título del gráfico
plt.grid(True)  # Activar la cuadrícula en el gráfico
plt.legend()  # Mostrar leyenda

# Mostrar la gráfica
plt.show()
````

#### Ajustando los límites de los ejes

**Matplotlib** hace un trabajo decente al elegir los límites de ejes predeterminados para su gráfico, pero a veces es bueno tener un control más preciso.  La forma más básica de ajustar los límites es utilizar las funciones `plt.xlim` y `plt.ylim`.

````python
import numpy as np
import matplotlib.pyplot as plt


# Datos para las tres gráficas
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Crear una nueva figura y ejes
fig=plt.figure(figsize=(8, 6))  # Tamaño de la figura
plt.ylim(-10,10) # Muestra el eje y de [-10,10]
plt.xlim(0,9) # Muestra el eje x de [0,9]
plt.plot(x, y1, '-b', label='sen(x)')  # Graficar seno
plt.plot(x, y2, '--r', label='cos(x)')  # Graficar coseno
plt.plot(x, y3, '-.g', label='tan(x)')  # Graficar tangente

# Personalizar la visualización
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Tres funciones trazadas en una gráfica')
plt.legend()  # Mostrar leyenda
plt.grid(True)  # Mostrar cuadrícula
plt.legend(loc='best', fontsize=10, frameon=True, shadow=True) # Personalizar la leyenda

# Mostrar la gráfica
plt.show()
#fig.savefig('my_figure.png')
````

### <font color='orange'> Creación de subgráficas en una figura </font>
--- 


`plt.subplot()` es una función de **Matplotlib** que se utiliza para crear y seleccionar subgráficas en una figura. Sintaxis:

```python 
plt.subplot(nrows, ncols, index)
```

Esta función toma tres argumentos:
- **nrows:** Número de filas de subgráficas en la figura.
- **ncols:** Número de columnas de subgráficas en la figura.
- **index:** Número de la subgráfica actual que se está creando o seleccionando, comenzando desde 1 y 
aumentando de izquierda a derecha y de arriba a abajo.

````python
import matplotlib.pyplot as plt
import numpy as np

# Datos para la primera gráfica
x = np.linspace(0, 4*np.pi, 100)
y1 = np.sin(x) #primera gráfica
y2 = np.cos(x) #segunda gráfica

# Crear una figura y dos subgráficas en una fila
plt.figure(figsize=(6, 3), dpi=100)  # Tamaño de la figura
#-----------------------------------
plt.subplot(1, 2, 1)  # Subgráfica 1
plt.plot(x, y1, color='blue')
plt.title('Gráfica 1: Seno')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)  # Mostrar cuadrícula

#-----------------------------------
plt.subplot(1, 2, 2)  # Subgráfica 2
plt.plot(x, y2, '--r')
plt.title('Gráfica 2: Coseno')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)  # Mostrar cuadrícula
#-----------------------------------
plt.tight_layout()  # Ajustar el espaciado
plt.show()
````

### <font color='blue'> Gráfico de dispersión (scatter plot) en Matplotlib </font>

---

La función `scatter()` en *Matplotlib* se utiliza para crear un **gráfico de dispersión**, que es una representación 
gráfica de puntos en un espacio bidimensional. Cada punto en el diagrama de dispersión representa una observación en un conjunto de datos 
y se define por su posición en el plano $XY$, donde cada eje representa una variable.

Sintaxis:
```python
plt.scatter(x, y, s=None, c=None, marker=None)

```
- x: Un array o lista de valores que representan las coordenadas $x$ de los puntos.
- y: Un array o lista de valores que representan las coordenadas $y$ de los puntos.

Los puntos se pueden personalizar utilizando:
- s: Tamaño de los puntos.
- c: Color de los puntos. Puede ser un color o una secuencia de colores. Por defecto, los puntos se representan en azul.
- marker: Tipo de marcador utilizado para representar los puntos.

Algunos tipos comunes de marcadores que puedes utilizar en *Matplotlib*:

- 'o': Círculo
- 's': Cuadrado
- '^': Triángulo hacia arriba
- 'v': Triángulo hacia abajo
- '>': Triángulo hacia la derecha
- '<': Triángulo hacia la izquierda
- '+': Cruz
- 'x': Cruz (alternativa)
- '*': Estrella
- '.': Punto
- ',': Píxel
- 'D': Diamante
- 'p': Pentágono
- 'H': Hexágono

````python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 4, 6]

x_1 = np.array([3, 8, 5, 6, 1, 9, 6, 7, 2, 1, 8])
y_1 = np.array([4, 5, 2, 4, 6, 1, 4, 6, 5, 3, 3])

# Crear el diagrama de dispersión
plt.scatter(x, y, s=40, c='red', marker='s')
plt.scatter(x_1, y_1, s=30, c='cyan', marker='*')

# Personalizar la visualización
plt.title('Diagrama de Dispersión Simple')
plt.xlabel('X')
plt.ylabel('Y')

# Mostrar el diagrama de dispersión
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.5, 3.5, 4.5, 5.5, 6.5])

# Graficar los datos con diferentes marcadores
plt.scatter(x, y, marker='o', label='Círculo')
plt.scatter(x, y + 1, marker='s', label='Cuadrado')
plt.scatter(x, y + 2, marker='^', label='Triángulo hacia arriba')
plt.scatter(x, y + 3, marker='v', label='Triángulo hacia abajo')
plt.scatter(x, y + 4, marker='>', label='Triángulo hacia la derecha')

# Agregar etiquetas y título
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Diferentes marcadores en scatter con NumPy arrays')

# Agregar una leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
````

#### Generar datos aleatorios con NumPy: Uso de `random`

````python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios
np.random.seed(0) # semilla del generador de números aleatorios de NumPy en 0.
#Se genera un array 'x' y 'y' de 50 números aleatorios distribuidos 
#uniformemente en el intervalo [0, 1).
x = np.random.rand(50) 
y = np.random.rand(50)

# Crear un diagrama de dispersión
plt.scatter(x, y, s=35, c='blue', marker='^', alpha=0.8)

# Personalizar la visualización
plt.title('Diagrama de Dispersión: Datos aleatorios')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')

# Mostrar el diagrama de dispersión
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Generar datos aleatorios
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(10, 100, size=50)
colors = np.random.rand(50)

# Crear un diagrama de dispersión
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)

# Personalizar la visualización
plt.title('Diagrama de Dispersión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.colorbar(label='Color')

# Mostrar el diagrama de dispersión
plt.show()
````
