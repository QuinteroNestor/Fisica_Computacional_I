## <font color='blue'> Histogramas en Matplotlib  </font>

---

Los **histogramas** son una herramienta poderosa para visualizar la distribución de datos numéricos. 
En **Matplotlib**, puedes crear histogramas utilizando la función `plt.hist()`.

### Características principales:

- **Visualización de la distribución**: Los histogramas muestran la frecuencia de ocurrencia de 
los valores en un conjunto de datos, lo que permite visualizar rápidamente la distribución de los datos.

- **Barras contiguas**: Los datos se agrupan en intervalos llamados "bins". 
Cada bin representa un rango de valores y la altura de la barra indica la frecuencia o número de ocurrencias en ese rango.

- **Personalización**: Matplotlib proporciona una amplia gama de opciones para personalizar la 
apariencia del histograma, incluyendo color de las barras, ancho de las barras, estilo de las líneas, 
etiquetas de ejes, títulos y más.

### Sintaxis básica:

```python
import matplotlib.pyplot as plt

# Crear un histograma
plt.hist(data, bins=None, range=None, density=False, color='blue', alpha=0.7 ...)

```
Toma una serie de datos `data` como entrada y puede aceptar varios otros argumentos opcionales, 
como el número de `bins`, el rango de los bins: `range`, si se desea normalizar la densidad, color, etc.

- El parámetro `bins` en la función `plt.hist()` especifica el número de contenedores (o intervalos) en los 
que se dividirá el rango de los datos para construir el histograma. 
En otras palabras, controla la cantidad de barras que se mostrarán en el histograma.

- El parámetro `density` en la función `plt.hist()` controla si se muestra el histograma de frecuencia (False) o el 
histograma de densidad de probabilidad (True). Esto es útil para comparar distribuciones con diferentes 
tamaños de muestra o diferentes rangos de datos.

## <font color='red'> Distribución normal (gaussiana) </font>

---

Una **distribución normal**, también conocida como **distribución gaussiana**, es una de las distribuciones de 
probabilidad más comunes en estadísticas y ciencias naturales. 
Se caracteriza por tener una forma de campana simétrica alrededor de su media, 
con la mayoría de los datos cerca de la media y una dispersión que disminuye a medida que se aleja de ella. 
La distribución normal está completamente determinada por dos parámetros: la media ($\mu$) y la desviación estándar ($\sigma$). 
La función de densidad de probabilidad de una distribución normal se define como: 

$$ f(x|\mu,\sigma) = \frac{1}{\sigma \sqrt{2\pi}} \, e^{-\frac{(x-\mu)^2}{2\sigma^2}}
 $$

La función `np.random.normal()` en **NumPy** se utiliza para generar muestras aleatorias de una distribución normal. 
Toma la media $\mu$, la desviación estándar $\sigma$ y el tamaño de la muestra como 
argumentos y devuelve un array de valores aleatorios que siguen una distribución normal.

### <font color='red'> Ejemplo: </font> Histogramas en análisis de datos

---

Los histogramas también son ampliamente utilizados en análisis de datos para explorar la 
distribución de una variable y detectar patrones o anomalías en los datos.

Supongamos que tenemos un conjunto de datos que representa las calificaciones de estudiantes en un examen. 
Podemos utilizar un histograma para visualizar la distribución de las calificaciones y 
comprender mejor cómo están distribuidas las notas en la clase.

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Generar datos de calificaciones de estudiantes (ejemplo ficticio)
calificaciones = np.random.normal(loc=70, scale=10, size=100)

# Crear histograma de calificaciones
plt.hist(calificaciones, bins=10, color='skyblue', edgecolor='black')

# Agregar etiquetas y título
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Distribución de calificaciones en un examen')

# Mostrar el histograma
plt.show()

````

En este ejemplo, generamos datos de calificaciones de estudiantes utilizando una distribución normal con una media de 70 y una desviación estándar de 10. Luego, creamos un histograma para visualizar la distribución de estas calificaciones. Esto nos permite analizar la distribución de las calificaciones en el examen y obtener información sobre el rendimiento de los estudiantes.

### <font color='red'> Ejemplo: </font> Histogramas en física de partículas

---

En física de partículas, los histogramas son herramientas fundamentales para analizar datos experimentales, 
especialmente para estudiar la distribución de energía, momento, masa invariante, entre otras variables.

Supongamos que estamos analizando datos de un experimento de colisiones de partículas en un acelerador de partículas. 
Después de la colisión, medimos la energía de las partículas resultantes. 
Para analizar estas mediciones, podemos utilizar un histograma para visualizar la distribución de energía de las partículas.

````python
# Generar datos de energía de partículas (ejemplo ficticio)
energias = np.random.normal(loc=100, scale=10, size=1000)

# Crear histograma de energías
plt.hist(energias, bins=30, color='darkgray', edgecolor='black')

# Agregar etiquetas y título
plt.xlabel('Energía (GeV)')
plt.ylabel('Frecuencia')
plt.title('Distribución de energía de partículas')

# Mostrar el histograma
plt.show()
````

En este ejemplo, generamos datos de energía de partículas utilizando una distribución normal con media de $\mu = 100$ GeV y
desviación estándar de $\sigma = 10$ GeV. Luego, creamos un histograma para visualizar la distribución de estas energías. 
Esto nos permite analizar la distribución de energía de las partículas producidas en las colisiones.

### <font color='red'> Ejemplo: </font> Varios histogramas en una sola figura

````python
# Generar datos de ejemplo para tres distribuciones
data1 = np.random.normal(loc=0, scale=1, size=1000)
data2 = np.random.normal(loc=2, scale=1.5, size=1000)
data3 = np.random.normal(loc=-2, scale=0.5, size=1000)

# Crear histogramas para cada conjunto de datos
plt.hist(data1, bins=30, color='skyblue', alpha=0.5, label='Datos 1')
plt.hist(data2, bins=30, color='salmon', alpha=0.5, label='Datos 2')
plt.hist(data3, bins=30, color='green', alpha=0.5, label='Datos 3')

# Agregar etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Comparación de distribuciones')
plt.legend()

# Mostrar el plot con los histogramas superpuestos
plt.show()
````

## Uso de la escala logarítmica

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios
x = np.linspace(1, 100, 200)
y = np.random.rand(200) * 1000

# Graficar los datos en escala logarítmica
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de dispersión en escala logarítmica')
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos para un gráfico de línea
x = np.linspace(0.1, 10, 100)
y = np.log(x)

# Graficar los datos en escala logarítmica
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico de línea en escala logarítmica')
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x
x = np.linspace(0.1, 10, 100)

# Función exponencial: y = e^x
y = np.exp(x)

# Graficar la función en escala lineal y logarítmica
plt.figure(figsize=(7, 5))

# Gráfico en escala lineal
plt.subplot(1, 2, 1)
plt.plot(x, np.exp(x))
plt.plot(x, np.exp(-x))

plt.title('Escala lineal')
plt.xlabel('x')
plt.ylabel('y')

# Gráfico en escala logarítmica
plt.subplot(1, 2, 2)
plt.plot(x, y, label='$e^{x}$')
plt.plot(x, np.exp(-x), label='$e^{-x}$')
plt.yscale('log')
plt.title('Escala logarítmica en y')
plt.xlabel('x')
plt.ylabel('y (log)')
plt.legend()

plt.tight_layout()
plt.show()
````

En este ejemplo, graficamos la función exponencial $ y(x) = e^x$ tanto en una escala lineal como en una escala logarítmica en el eje y. Como la función exponencial crece rápidamente, usar una escala logarítmica puede ayudar a visualizar mejor la función en un rango más amplio de valores de y.

````python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x
x = np.linspace(0.1, 10, 100)

# Función de decaimiento exponencial: y = e^(-x)
y = np.exp(-x)

# Graficar la función en escala lineal y logarítmica
plt.figure(figsize=(10, 6))

# Gráfico en escala lineal
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title('Escala lineal')
plt.xlabel('x')
plt.ylabel('y')

# Gráfico en escala logarítmica
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.yscale('log')
plt.title('Escala logarítmica en y')
plt.xlabel('x')
plt.ylabel('y (log)')
plt.tight_layout()
plt.show()
````
